```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, num=100, endpoint=False)
```

```python
m, c = 10, 2
y = m * (x**2) + c
```

```python
y
```

```python
plt.figure(figsize=(15, 10)) #width height
plt.plot(x, m*x**2 + c, label="Parabola", c=(3/255, 134/255, 161/255))
# plt.show()
plt.plot(x, m*x + c, label="Line", c=(3/255, 255/255, 162/255))
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("My Graph")
plt.legend()
plt.savefig("./my_graph.jpg")
plt.show()
```

```python
plt.figure(figsize=(15, 10)) #width height
plt.scatter(x, m*x**2 + c, label="Parabola", marker="<", color=(3/255, 134/255, 161/255))
# plt.show()
plt.scatter(x, m*x + c, label="Line", marker="*",color=(3/255, 255/255, 162/255))
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("My Graph")
plt.legend()
plt.savefig("./my_graph_1.jpg")
plt.show()
```
