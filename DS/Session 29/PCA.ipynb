```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
X1, X2 = make_regression(n_features=1, noise=20)
```

```python
X1.shape, X2.shape
```

```python
plt.scatter(X1, X2)
plt.show()
```

```python
X = np.hstack((X1.reshape(-1,1), X2.reshape(-1,1)))
```

```python
X = (X - X.mean(axis=0))/X.std(axis=0)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
eigen_val, eigen_vector = np.linalg.eig(X.T.dot(X))
```

```python
eigen_vector.shape
```

```python
eigen_vector
```

```python
eigen_vector[:, 0], eigen_val[0]
```

```python
eigen_vector[:, 0].T.dot(eigen_vector[:, 1])
```

```python
pc1 = eigen_vector[:, np.argsort(eigen_val)[-1]]
pc2 = eigen_vector[:, np.argsort(eigen_val)[-2]]
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.plot([0, pc1[0]], [0, pc1[1]], c="red", label="principal component 1")
plt.plot([0, pc2[0]], [0, pc2[1]], c="green", label="principal component 2")
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X)
```

```python
X_transformed = pca.transform(X)
```

```python
plt.scatter(X_transformed[:, 0], X_transformed[:, 1])
plt.show()
```
